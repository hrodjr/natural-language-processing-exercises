```python
from requests import get
from bs4 import BeautifulSoup
import os
import re
import unicodedata
import pandas as pd
import nltk
from nltk.tokenize.toktok import ToktokTokenizer
from nltk.corpus import stopwords
from speeches import acquire_speech, clean_speech, acquire_speech_avalon
from wordcloud import WordCloud
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## George Washington 1789

```python
url = 'https://avalon.law.yale.edu/18th_century/wash1.asp'
words = acquire_speech_avalon(url)
speech = clean_speech(words)
img = WordCloud(background_color='white').generate(speech)
# WordCloud() produces an image object, which can be displayed with plt.imshow
plt.imshow(img)
# axis aren't very useful for a word cloud
plt.axis('off')
```

## Donald Trump 2017

```python
url = 'https://trumpwhitehouse.archives.gov/briefings-statements/the-inaugural-address/'
response = get(url, headers = {'User-Agent': 'Codeup Data Science'})
soup = BeautifulSoup(response.text, features='lxml')
speech = soup.select('.page-content')
speech = speech[0].select('p')
words = [words.text for words in speech]
speech = clean_speech(words)
img = WordCloud(background_color='white').generate(speech)
# WordCloud() produces an image object, which can be displayed with plt.imshow
plt.imshow(img)
# axis aren't very useful for a word cloud
plt.axis('off')
```
