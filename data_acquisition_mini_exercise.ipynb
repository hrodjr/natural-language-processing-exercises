```python
import pandas as pd
from requests import get
from bs4 import BeautifulSoup
import os
```

```python
url = 'https://web-scraping-demo.zgulde.net/people'
response = get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
people = soup.select('.person')
```

```python
person = people[0]
```

```python
#name
person.h2.text
```

```python
#quote
person.p.text.strip()
```

```python
#email
person.select('.email')[0].text
```

```python
#phone
person.select('.phone')[0].text
```

```python
#address
address = person.select('.address')[0].text.strip()
address
```
